
# Draw histogram

```python
data = np.random.randn(10000)
plt.figure(figsize=(10,10))
plt.title("random number histogram",fontsize=22 )
plt.xlabel("rnd number ", fontsize = 20)
plt.ylabel("frequency ", fontsize = 20)
plt.grid(True) # or axis = 'x'
plt.text(-3.5, 600, r'$\mu=0,\ \sigma=1$, sarkar',fontsize=12,color='k')
plt.hist(data,50, color = 'red', alpha =0.3)


#fitfunc  = lambda p, x: p[0]*exp(-0.5*((x-p[1])/p[2])**2)+p[3]
#errfunc  = lambda p, x, y: (y - fitfunc(p, x))
#out   = leastsq( errfunc, init, args=(xdata, ydata))
```

```python
# Multiple function in same graph
x = np.arange(2,6,1)
y = [(lambda k: k*k)(k) for k in x]
z = [(lambda k: k*k*k)(k) for k in x]
```

```python
plt.plot(x, y,color='red',linewidth=3.0,label='x*x',ls='--')
plt.plot(x,z,color='green', label='x*x*x')
plt.legend()
plt.savefig('twoplot.png')
```
